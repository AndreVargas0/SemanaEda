```python
import pandas as pd
```

```python
pesquisa = pd.read_csv('dataset/Pesquisa.csv')
```

```python
pesquisa.columns = ['data_hora','rock','hiphop','mpb','reggae','funk','sertanejo','gospel','eletronica']
```

```python
df = pesquisa.drop(columns=['data_hora'])
```

```python
soma_pontuacoes = df.sum()
```

```python
soma_pontuacoes.plot(kind='barh')
```

```python
df_estilos = soma_pontuacoes.reset_index()

df_estilos.columns = ['estilo', 'pontuacao']

print(df_estilos)
```

```python
df_estilos['pontuacao'].sum()
```

```python
dados_grafico = df_estilos.groupby('estilo').sum()['pontuacao'].sort_values()
```

```python
dados_grafico.plot(kind='hist')
```
